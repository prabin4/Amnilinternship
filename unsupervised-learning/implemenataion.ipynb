IMplemetation K means and DB scan


```python
from sklearn.datasets import load_wine
import pandas as pd
from sklearn.preprocessing import StandardScaler

wine_data = load_wine(as_frame=True)
X = wine_data.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.info()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')   
kmeans.fit(X_scaled_df)
kmeans_labels = kmeans.labels_

plt.figure(figsize=(10, 6))
plt.scatter(X_scaled_df.iloc[:, 0], X_scaled_df.iloc[:, 1], c=kmeans_labels, cmap='viridis', marker='o')  
plt.title('K-Means Clustering of Wine Data')
plt.xlabel(X_scaled_df.columns[0])
plt.ylabel(X_scaled_df.columns[1])
plt.colorbar(label='Cluster Label')
plt.show()
```

DB scan implementation


```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris

iris= load_iris(as_frame=True)
X = iris.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.info()


dbscan = DBSCAN(eps=0.8, min_samples=15)
dbscan.fit(X_scaled_df)
dbscan_labels = dbscan.labels_

plt.figure(figsize=(10, 6))
plt.scatter(X_scaled_df.iloc[:, 0], X_scaled_df.iloc[:, 1], c=dbscan_labels, cmap='viridis', marker='o')
plt.title('DBSCAN Clustering of Wine Data')
plt.xlabel(X_scaled_df.columns[0])
plt.ylabel(X_scaled_df.columns[1])
plt.colorbar(label='Cluster Label')
plt.show()
```
